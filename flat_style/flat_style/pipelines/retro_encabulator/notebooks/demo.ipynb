# Flat-layout Retro-Encabulator demo

This notebook demonstrates a direct import from the `flat_style` package. It requires the `flat_style` project root to be available as a Python import root; it does not install a package or modify `sys.path`.

```python
from flat_style.pipelines.retro_encabulator.instruments import (
    InverseReactiveCurrent,
    RetroEncabulator,
    synchronize_cardinal_gram_meters,
)

current = InverseReactiveCurrent(ohms=42.0)
encabulator = synchronize_cardinal_gram_meters(RetroEncabulator())

print(f'Inverse reactive current: {current.ohms} ohms')
print(encabulator.stabilize_milford_trunions())
```
